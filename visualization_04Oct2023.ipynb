```python
days = tips_df.day.unique()
print(days)

day_data = {}

for uni_day in days:
  day_data[uni_day] = [tips_df[tips_df['day']==uni_day]['day'].count()]

print(day_data)
```

```python
day_df = pd.DataFrame(day_data)
sns.barplot(day_df)
```
